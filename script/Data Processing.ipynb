```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv(r"D:\Lecture Note\Investigation (FYP)\bitext_free_dataset.csv")
dataframe = pd.DataFrame(data)

print(dataframe['utterance'])
               
```

```python
intent = []
```

```python
for x in dataframe['intent']:
    if x not in intent:
        intent.append(x)

intent_list = np.unique(intent)
print(intent_list)
```

```python
table = dict()

for word in intent_list:
    if word not in table:
        table.append(word)


```

```python
for word in intent_list:
    for i in range(len(dataframe['utterance'])):
        if dataframe.iloc[i]['intent'] != word:
            continue
        else:
            if word in table:
                table[word].append(dataframe.iloc[i]['utterance'])
            else:
                table[word] = [dataframe.iloc[i]['utterance']]
##table.update({word:dataframe.iloc[i]['utterance']})
```

```python
print(table)
```

```python
result = pd.DataFrame({key:pd.Series(value) for key, value in table.items()})
```

```python
def create_rasa_files(path, create_files_path):
    """
    Converts an CSV file created with the specified format to RASA accepted nlu.md format
    :param path: path where the CSV file is present
    :param create_files_path: path where the nlu.md file needs to be created
    :return: return nothing. A file is created in the path specified via create_files_path
    """
    df = pd.read_csv(r"{}".format(path))
    file = open(r'{}\nlu.md'.format(create_files_path), "w")
    intents = list(df.columns)
    for item in intents:
        file.write("## intent: {intent_name}\n".format(intent_name=item))
        for sent in df[item]:
            file.write("- {}\n".format(sent))
    file.close()
    file = open(r'{}\domain.yml'.format(create_files_path), "w")
    file.write("intents:\n")
    for intent_name in intents:
        file.write("  - {}\n".format(intent_name))
    file.write("actions:\n")
    for intent_name in intents:
        file.write("  - utter_{}\n".format(intent_name))
    file.close()
    return None
```
